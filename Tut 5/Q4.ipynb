```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression   
```

```python
training_students = np.array([36,26,35,39,26,30,31,38,36,38,26,26])
training_books = np.array([31,20,34,35,20,30,30,38,34,33,20,20])


model = LinearRegression()
model.fit(training_books.reshape(-1,1), training_students)

print(f"Regression coefficients: {model.coef_}")
```

```python
print(model.predict([[30]]))
```

```python

unique_data = np.unique(np.column_stack((training_books, training_students)), axis=0)
students_unique = unique_data[:,1].reshape(-1,1)
books_unique = unique_data[:,0]


model_without_duplicates = LinearRegression()
model_without_duplicates.fit(students_unique, books_unique)




print(model_without_duplicates.predict([[30]]))
```

```python
plt.figure()
plt.scatter(training_students, training_books,color='blue',label='Data with duplicates')
plt.scatter(students_unique, books_unique,color='red',label='Data without duplicates')  
plt.plot(training_students, model.predict(training_students.reshape(-1,1)), color='green', label='Linear Regression with duplicates')
plt.plot(students_unique, model_without_duplicates.predict(students_unique), color='orange', label='Linear Regression without duplicates')
plt.xlabel('Number of students')
plt.ylabel('Number of books')
plt.legend()
plt.show()
```
