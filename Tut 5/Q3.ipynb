```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

Obtain a scatter plot of the number of books sold versus the number of registered students

```python
number_of_students = np.array([36,28,35,39,30,30,31,38,36,38,29,26])
number_of_books = np.array([31,29,34,35,29,30,30,38,34,33,29,26])


plt.scatter(number_of_students, number_of_books)
plt.xlabel("Number of students")
plt.ylabel("Number of books")
plt.show()
```

```python
model = LinearRegression()
model.fit(number_of_students.reshape(-1,1), number_of_books)

print(f"Regression equation: y = {model.coef_[0]}x + {model.intercept_}")
print(f"Coefficient: {model.coef_[0]}")
```

```python
test_students = np.array([30])
predicted_books = model.predict(test_students.reshape(-1,1))
print(f"Predicted number of books: {predicted_books[0]}")
```
