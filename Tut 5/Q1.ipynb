```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X = np.array([-10,-8,-3,-1,2,8]).reshape(-1,1)
Y = np.array([5,5,4,3,2,2])

linear_regression_model = LinearRegression(fit_intercept=True)
linear_regression_model.fit(X, Y)
y_pred = linear_regression_model.predict(X)

linear_reegression_model_without_bias = LinearRegression(fit_intercept=False)
linear_reegression_model_without_bias.fit(X, Y)
y_pred_without_bias = linear_reegression_model_without_bias.predict(X)

plt.figure()

plt.subplot(1,2,1)
plt.scatter(X, Y,color='blue',label='Data')
plt.plot(X, y_pred, color='red', label='Linear Regression with bias')
plt.xlabel('X')
plt.ylabel('Y')

plt.subplot(1,2,2)
plt.scatter(X, Y,color='blue',label='Data')
plt.plot(X, y_pred_without_bias, color='red', label='Linear Regression without bias')
plt.xlabel('X')
plt.ylabel('Y')


plt.show()
```

Adding a bias allows the model to shift the fitted line up and down without forcing to go through the origin
