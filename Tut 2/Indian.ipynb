```python
import pandas as pd
import numpy as np
```

```python
indian = pd.read_csv('Indian.csv')
indian.describe()
```

```python
#Count the number of 0 entries in columns 1,2,3,4,5

indian_set = indian[0:5]
count_zero = indian_set[indian_set == 0].count()
print(count_zero)
```

```python
#Replace with nan

indian_set = indian[0:5]
indian_set.replace(0, np.nan, inplace=True)
print(indian_set)
count_zero = indian_set[indian_set == 0].count()
print(count_zero)
```
